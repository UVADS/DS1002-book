# Brief introduction to programming

Computers execute instructions to process data. The way we speak with a computer is through a programming language, which, like any other language, contains rules about spelling and grammar.  This course we will focus on Python  and R (briefly), but there are many other languages.

## Compiled vs Interpreted languages

Depending on how languages communicate with the computer before executing the provided instructions, we can find two kinds of programming languages. In a **compiled** language, our whole code is first translated (=compiled) and this translation is stored into a different executable file. Examples of compiled languages are Fortran, C and C++. In contrast, we have other **interpreded** languages, which translate and execute code line-by-line. Python and R are examples of this.

A compiled language comes with a compiler, which compiles source files to executable binary files. An interpreted language comes with an interpreter, which interprets source files and executes them. This interpretation tends to make programming languages like Python and R less efficient than compiled languages. That is why many programs (e.g. video games) are written in a language like C, because they need to be efficient in the use of the computer resources. They do this by an explicit management of the available resources (e.g. memory). For this reason, it is often said that in compiled languages one programmes at a lower level than in languages like Python, where there is an interpreter that does a lot of work for you.

In data science (and science in general), unless your research question needs to be resource-intesive, languages like Python and R work perfect great, because they allow you to address your research question easily and rapidly. In addition, speed difference on a today's computers is usually negligible. Finally, code in higher-level language can me more readable, which makes your applications easier to understand by others.

- Procedural Languages: These languages follow a set of instructions or procedures to solve a problem. They are generally easy to understand and write. Examples include C, Fortran, and Pascal. 

- Object-Oriented Languages: These languages are based on the concept of "objects," which can contain data and code to manipulate that data. They help organize complex programs and promote code reusability.Examples include Java, C++, and Python. 

- Functional Languages: These languages are based on mathematical functions and avoid changing states or mutable data. They are often used in applications that require high levels of mathematical computations. Examples include Lisp, Haskell, and Scala.

- Scripting Languages: These are typically used for automating tasks and, in contrast to procedural languages, they are often interpreted rather than compiled. They are commonly used for web development, system administration, and quick prototyping. Examples include JavaScript, Python, Bash, and Ruby. 

- Logic Programming Languages: These languages are based on formal logic and are used primarily in artificial intelligence and problem-solving applications. Prolog is a well-known example.

As you can see, Python appears in two of these categories. 

In Python, **everything** is an object However, it can be used to write code in multiple styles â€“ procedural, object-oriented or even functional.

## Programming and Data Science

Both R and Python, along with SQL; not covered in this course, are the most widely used programming languages for Data Science.

Let's have a look at some of their features:




You can have a look at some numbers from the 2022 Kaggle survey (https://www.kaggle.com/competitions/kaggle-survey-2022):

```python
# Import libraries to read data and visualize
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt

# Read survey data
survey_data = pd.read_csv("../../data/kaggle_survey_2022_responses.csv", low_memory=False) 
# Count number of responses per category (programming language)
languages_dat = survey_data.filter(regex="Q12").apply(lambda x: x[1:].value_counts()).fillna(0).sum(axis=1)
# Get number of responses
n_obs = survey_data.shape[0]
# Accomodate data for plotting
languages_dat = pd.DataFrame({"language": languages_dat.index, "use": 100*languages_dat/n_obs})
languages_dat = languages_dat.reset_index(drop=True)

# Plot results
sns.barplot(data=languages_dat, y="language", x="use", orient="h")
plt.tick_params(labelsize=15)
plt.xlabel("Use (%)", size=20)
plt.ylabel("")
sns.despine()
```

As you can see, Python beats R numbers in terms of use. Why is that?

- Python is a multi-paradigm/multi-purpose language, which means that can be used for a wide variety of applications beyond data analysis, such as web development, automation, and software development. R, in contrast, was designed specifically for statistical analysis, so its use and development has mostly focused on this area.
- Even though both languages may have an equivalent learning curve for beginners, those wih previous experience in programming will feel Python more intuitive than R.
- Although both both are widely used in academia, its versatility  and better suitability for production makes Python the first choice in data science related application in industry.
- Related to the previous point, doing machine and deep learning feels way easier in Python. Scikit-learn (https://scikit-learn.org/stable/) for machine learning, and Tensorflow (https://www.tensorflow.org/) and pytorch(https://pytorch.org/) for deep learning, are probably the most used libraries for machine and deep learning, and both are written in Python.
